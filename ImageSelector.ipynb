```python
# Importing prerequisites
import jsonlines
import shutil
```

```python
# Selecting source folder and destination folder
source = "D:/download/Downloads/pubtabnet/pubtabnet/train/"
destination = "datasetImages/"
```

```python
# Loading an example annotation and extracting imaegs
with jsonlines.open('mariani.jsonl', 'r') as reader:
    for line in reader:
        filename = line['filename']
        try:
            # Sposta il file
            shutil.move(source+filename, destination+filename)
            print(f"File {filename} moved successfully from {source} to {destination}.")
        except FileNotFoundError:
            print(f"File {filename} not found in {source}.")
        except FileExistsError:
            print(f"a file with the same name already exists in {destination}.")
```
