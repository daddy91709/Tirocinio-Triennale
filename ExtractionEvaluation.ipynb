```python
# import and setup
import os
from metric import TEDS

# Initialize TEDS object
teds = TEDS()

ground_truth = "HTML/ground truth/"
extracted = "HTML/extracted/"

#contatori
max=0
tot=0
i=0
compared=0
```

```python
# data extraction
files = os.listdir(extracted)

for file in files:
    with open(extracted+file+'/res_0.txt', "r", encoding="utf-8") as content:
        text = content.read()
        #select html text only and clean it
        html_to_test = text[text.index('<'):-18].replace('"','').replace('\\','')

        with open(ground_truth+file+'.txt', "r", encoding="utf-8") as true_html:
            i= i+1
            # Evaluate
            try:
                score = teds.evaluate(html_to_test, true_html.read())

                result = str(i)+") file: "+file+" -score[0,1]: "+str(score)+" -score(%): "+str(score*100)

                #max & avg score
                compared= compared+1
                tot = tot+score
                if score>max : max = score

            except:
                result = "cannot evaluate score for file "+file

            print(result + '\n')

            with open('scores.txt', "a", encoding="utf-8") as file:
                file.write(result + '\n')

with open('scores.txt', "a", encoding="utf-8") as file:
    file.write('\n'+'-------------------------------------------------'+'\n'+'-max: '+str(max)+' -avg: '+str(tot/compared))
```
